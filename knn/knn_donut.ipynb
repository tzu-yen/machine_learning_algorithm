```python
import numpy as np
import matplotlib.pyplot as plt
from knn import KNN
%matplotlib inline
```

```python
def get_donut_data():
    N = 200
    R_in, R_out = 5, 10
    R1 = R_in + np.random.random(N/2)
    theta1 = 2 * np.pi * np.random.random(N/2)
    X_inner = np.concatenate([[R1*np.cos(theta1)], [R1*np.sin(theta1)]]).T

    R2 = R_out + np.random.random(N/2)
    theta2 = 2 * np.pi * np.random.random(N/2)
    X_outer = np.concatenate([[R2*np.cos(theta2)], [R2*np.sin(theta2)]]).T
    X = np.concatenate([X_inner, X_outer])
    y = np.array([0]*(N/2) + [1]*(N/2))

    plt.scatter(X[:,0], X[:,1], c=y, s=100, alpha=0.5)
    return X, y
```

```python
if __name__ == '__main__':
    X, y = get_donut_data()
    model = KNN(3)
    model.fit(X, y)
    print 'Training Accuracy:', model.score(X, y)
```
